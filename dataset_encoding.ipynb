```python
import pandas
import numpy
import itertools
import collections
```

```python
df = pandas.read_csv('IMDB-Dataset-clean-reduced.csv')
df.head(10)
```

```python
df['review'] = df['review'].apply(str.split)
df.head(10)
```

```python
word_set = list(set(itertools.chain.from_iterable(df['review'])))
word_encodings = dict((word, index) for index, word in enumerate(word_set))
df['review'] = df['review'].apply(lambda words: list(map(lambda word: word_encodings[word], words)))
df.head(10)
```

```python
df['review'] = df['review'].apply(collections.Counter)
columns = dict((word_code, df['review'].apply(lambda counter: counter[word_code])) for word_code in range(len(word_set)))
def encode_sentiment(sent):
    if sent == 'positive':
        return 1
    elif sent == 'negative':
        return 0
    else:
        return -1
df['sentiment'] = df['sentiment'].apply(encode_sentiment)
```

```python
new_df = pandas.DataFrame(columns)
new_df = new_df.assign(sentiment = df['sentiment'])
new_df.head(10)
```

```python
new_df.to_csv("IMDB-reduced-encoded.csv")
```
