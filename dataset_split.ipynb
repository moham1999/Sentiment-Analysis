```python
import pandas
import sklearn.model_selection
```

```python
df = pandas.read_csv('IMDB-Dataset-clean-reduced.csv')
df.head(10)
```

```python
df_train, df_nontrain = sklearn.model_selection.train_test_split(df, train_size=0.8, shuffle=True, random_state=1337)
df_valid, df_test = sklearn.model_selection.train_test_split(df_nontrain, train_size=0.5, shuffle=False, random_state=1337)
```

```python
df_train.to_csv("IMDB-Dataset-training.csv", index=False)
df_train.head(10)
```

```python
df_valid.to_csv("IMDB-Dataset-validation.csv", index=False)
df_valid.head(10)
```

```python
df_test.to_csv("IMDB-Dataset-testing.csv", index=False)
df_test.head(10)
```
